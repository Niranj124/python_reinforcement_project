replacing the outliers using the IQR method

```python
numeric_col=df.select_dtypes(include=['number'])
for i in range(10):
   for col in numeric_col.columns:
       Q1 = df[col].quantile(0.25)
       Q3 = df[col].quantile(0.75)
       IQR = Q3 - Q1
        
       lower_bound = Q1 - 1.5* IQR
       upper_bound = Q3 + 1.5* IQR
        
       print(f"Lower bound:{lower_bound}")
       print(f"Upper bound:{upper_bound}")
       #calculate the median
       median_value = df[col].median()
       print(f"median value:{median_value}")
       #replace outliers
       df.loc[(df[col]<=lower_bound) | (df[col]>=upper_bound),col]=round(median_value,0)
       print("after replacing outliers:",df)
```

6.2.1    one way T-test:price

```python
from scipy.stats import ttest_1samp

# Drop missing values from the price column
sample_prices = df['price'].dropna()

# Define the hypothesized population mean (e.g., â‚¹500,000)
hypothesized_mean = 500000

# Perform one-sample t-test
t_stat, p_value = ttest_1samp(sample_prices, hypothesized_mean)

# Print test results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis.")
    print("Conclusion: The average house price is significantly different from â‚¹500,000.")
else:
    print("Fail to reject the null hypothesis.")
    print("Conclusion: There is no significant difference between the average house price and â‚¹500,000.")
```

6.2    two way T-Test: price vs is_renovated

Hypothesis:

Hâ‚€ (Null): Renovation has no effect on price

Hâ‚ (Alt): Renovated homes have a different mean price




```python
from scipy.stats import ttest_ind

# Split data
renovated = df[df['is_renovated'] == 1]['price']
not_renovated = df[df['is_renovated'] == 0]['price']

# Perform t-test 
t_stat, p_val = ttest_ind(renovated, not_renovated, equal_var=False)

# Significance level
alpha = 0.05

# Print results
print(f"T-Statistic: {t_stat:.2f}")
print(f"P-Value: {p_val:.5f}")

# Interpret results
if p_val < alpha:
    print(f"Since p-value ({p_val:.5f}) < alpha ({alpha}), we reject the null hypothesis.")
    print("Conclusion: There is a significant difference in mean price between renovated and not renovated homes.")
else:
    print(f"Since p-value ({p_val:.5f}) >= alpha ({alpha}), we fail to reject the null hypothesis.")
    print("Conclusion: No significant difference in mean price between renovated and not renovated homes.")
```

If p < 0.05, we reject Hâ‚€ â†’ renovation significantly affects price

If p >= 0.05, we fail to reject Hâ‚€ â†’ difference not statistically significant

 6.3    One way ANOVA test: price vs condition

Hypothesis:

Hâ‚€: Mean price is the same across all conditions

Hâ‚: At least one condition level has different mean price

```python
print(df['condition'].value_counts())
```

```python
from scipy.stats import f_oneway

# Group price by condition
grouped_prices = [df[df['condition'] == c]['price'].dropna() for c in df['condition'].unique()]

# Perform one-way ANOVA
f_stat, p_value = f_oneway(*grouped_prices)

# Print results
print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis.")
    print("Conclusion: At least one property condition group has a significantly different average price.")
else:
    print("Fail to reject the null hypothesis.")
    print("Conclusion: No significant difference in average price across property conditions.")
```

The ANOVA test showed a significant difference in prices across property conditions.
Better condition homes tend to have higher average prices, affecting buyer decisions.

6.4 Chi-Square Test â€“ Renovation Status vs Property Condition

```python
from scipy.stats import chi2_contingency

# Create contingency table
contingency_table = pd.crosstab(df['is_renovated'], df['condition'])

# Perform Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Output results
print("Chi-Square Statistic:", round(chi2, 2))
print("Degrees of Freedom:", dof)
print("P-Value:", round(p, 5))

if p < 0.05:
    print("Reject the null hypothesis - variables are not independent")
else:
    print("Fail to reject the null hypothesis - variables are independent")
```

Hâ‚€ (Null Hypothesis): Renovation status and condition rating are independent.

Hâ‚ (Alternative Hypothesis): Renovation status and condition rating are associated.


A low p-value (< 0.05) would suggest that renovated homes tend to have better condition ratings, which aligns with real-world expectations
Renovation status and condition are strongly associated
